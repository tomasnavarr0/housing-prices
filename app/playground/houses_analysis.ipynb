```python
from sklearn.model_selection import train_test_split
```

```python
final_dataset = dataset[["LotArea", "LotShape", "LotConfig", "Neighborhood", "OverallQual", "OverallCond", "YearBuilt", "GarageArea", "PoolArea", "BedroomAbvGr", "SaleCondition", "TotalBath", "OutdoorArea", "TotalArea", "SalePrice"]]
```

```python
categorical_cols = ['LotShape', 'LotConfig', 'Neighborhood', 'SaleCondition']
```

```python
final_dataset_encoded = pd.get_dummies(final_dataset, columns=categorical_cols, prefix=categorical_cols, drop_first=True)
```

```python
X = final_dataset_encoded.drop(columns=["SalePrice"])
y = final_dataset_encoded.SalePrice
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
import pandas as pd
import numpy as np
from sklearn.base import BaseEstimator
from sklearn.model_selection import cross_validate
from sklearn.metrics import (
    mean_absolute_error, 
    mean_squared_error, 
    r2_score, 
    mean_absolute_percentage_error,
    make_scorer
)

def get_metrics(y_test: pd.Series, y_pred: np.ndarray) -> None:
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    
    print(f"MAE {round(mae, 2)}")
    print(f"MSE {round(mse, 2)}")
    print(f"R2 {round(r2, 2)}")
    print(f"MAPE {round(mape, 2)}")

def train_model(model: type[BaseEstimator]) -> None:
    # Instanciar el modelo
    model_instance = model()
    
    # ValidaciÃ³n cruzada ANTES del entrenamiento final
    print("=== VALIDACIÃ“N CRUZADA ===")
    scorers = {
        'MAE': 'neg_mean_absolute_error',
        'MSE': 'neg_mean_squared_error', 
        'R2': 'r2',
        'MAPE': make_scorer(mean_absolute_percentage_error, greater_is_better=False)
    }
    
    cv_results = cross_validate(
        model_instance, X_train, y_train,
        cv=5,
        scoring=scorers
    )
    
    # Mostrar mÃ©tricas de CV
    for metric_name, scores in cv_results.items():
        if metric_name.startswith('test_'):
            clean_name = metric_name.replace('test_', '')
            if clean_name in ['MAE', 'MSE', 'MAPE']:
                cv_scores = -scores  # Convertir a positivo
            else:
                cv_scores = scores
            mean_score = np.mean(cv_scores)
            std_score = np.std(cv_scores)
            print(f"CV {clean_name}: {round(mean_score, 2)} (+/- {round(std_score * 2, 2)})")
    
    print("\n=== MÃ‰TRICAS EN TEST SET ===")
    
    # Tu lÃ³gica original
    model_instance.fit(X_train, y_train)
    y_pred = model_instance.predict(X_test)
    
    return get_metrics(y_test, y_pred)
```

```python
from sklearn.linear_model import LinearRegression

train_model(LinearRegression)
```

```python
from sklearn.linear_model import ElasticNet, ElasticNetCV

print("*****Elasticnet*****")
train_model(ElasticNet)
print("")
print("****ElasticNetCV*****")
train_model(ElasticNetCV)
```

```python
from sklearn.linear_model import Lasso, LassoCV

print("*****Lasso*****")
train_model(Lasso)
print("")
print("****LassoCV*****")
train_model(LassoCV)
```

```python
from sklearn.ensemble import (
    RandomForestRegressor, 
    GradientBoostingRegressor, 
    ExtraTreesRegressor,
    AdaBoostRegressor,
    BaggingRegressor
)

print("*****RandomForestRegressor*****")
train_model(RandomForestRegressor)
print("")
print("*****GradientBoostingRegressor*****")
train_model(GradientBoostingRegressor)
print("")
print("*****ExtraTreesRegressor*****")
train_model(ExtraTreesRegressor)
print("")
print("*****AdaBoostRegressor*****")
train_model(AdaBoostRegressor)
print("")
print("*****BaggingRegressor*****")
train_model(BaggingRegressor)
print("")
```

```python
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

print("*****SVR*****")
train_model(SVR)
print("")
print("*****KNeighborsRegressor*****")
train_model(KNeighborsRegressor)
print("")
print("*****DecisionTreeRegressor*****")
train_model(DecisionTreeRegressor)
print("")
```

```python
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5],
    'criterion': ['squared_error', 'absolute_error']
}

# GridSearch
dt = DecisionTreeRegressor(random_state=42)
mape_scorer = make_scorer(mean_absolute_percentage_error, greater_is_better=False)
grid = GridSearchCV(dt, param_grid, cv=3, scoring=mape_scorer, n_jobs=-1)

grid.fit(X, y)

#  Resultados
print("Mejores parÃ¡metros:")
print(grid.best_params_)
print(f"\nMejor MAPE: {-grid.best_score_:.3f} ({-grid.best_score_*100:.1f}%)")
```
