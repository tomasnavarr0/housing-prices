```python
def dataset_preprocessing(dataframe: pd.DataFrame) -> pd.DataFrame:
    dataframe.set_index("Id", inplace=True)
    dataframe.drop(columns=["PoolQC", "MiscFeature", "Alley", "Fence", "MasVnrType", "FireplaceQu", "LotFrontage"], inplace=True)
    dataframe.drop(columns=["MSZoning", "Street", "LandContour", "Utilities", "LandSlope", "Condition2", "RoofMatl", "BsmtCond", "BsmtFinType2", "Heating", "CentralAir", "Electrical", "LowQualFinSF", "Functional", "GarageFinish", "GarageQual", "GarageCond", "PavedDrive", "SaleType"], inplace=True)
    dataframe["TotalBsmtSF"] = dataframe["TotalBsmtSF"]-dataframe["BsmtUnfSF"]
    dataframe.drop(columns=["BsmtFinSF1", "BsmtFinSF2", "BsmtUnfSF"], inplace=True)
    dataframe["2ndFlrSF"] = dataframe["2ndFlrSF"].apply(lambda x:True if x > 0 else False)
    dataframe["2ndFlrSF"] = dataframe["2ndFlrSF"].astype(int)
    dataframe.drop(columns=["1stFlrSF"],inplace=True)
    dataframe["TotalBath"] = dataframe["FullBath"]+dataframe["HalfBath"]
    dataframe.drop(columns=["BsmtFullBath", "BsmtHalfBath", "FullBath", "HalfBath"],inplace=True)
    dataframe["OutdoorArea"] = dataframe["WoodDeckSF"] + dataframe["OpenPorchSF"] + dataframe["EnclosedPorch"] + dataframe["3SsnPorch"] + dataframe["ScreenPorch"]
    dataframe.drop(columns=["WoodDeckSF", "OpenPorchSF", "EnclosedPorch", "3SsnPorch", "ScreenPorch"],inplace=True)
    dataframe.drop(columns=["GarageType", "GarageYrBlt"],inplace=True)
    dataframe["TotalArea"] = dataframe["GrLivArea"]+dataframe["TotalBsmtSF"]+dataframe["GarageArea"]+dataframe["OutdoorArea"]
    dataframe.drop(columns=["HouseStyle", "BldgType"], inplace=True)
    final_dataframe = dataframe[["LotArea", "LotShape", "LotConfig", "Neighborhood", "OverallQual", "OverallCond", "YearBuilt", "GarageArea", "PoolArea", "BedroomAbvGr", "SaleCondition", "TotalBath", "OutdoorArea", "TotalArea"]]
    categorical_cols = ['LotShape', 'LotConfig', 'Neighborhood', 'SaleCondition']
    final_dataframe_encoded = pd.get_dummies(final_dataframe, columns=categorical_cols, prefix=categorical_cols, drop_first=True)
    return final_dataframe_encoded
```

```python
from typing import Any
import pandas as pd
import numpy as np
from sklearn.base import BaseEstimator
from sklearn.model_selection import cross_validate
from sklearn.metrics import (
    mean_absolute_error, 
    mean_squared_error, 
    r2_score, 
    mean_absolute_percentage_error,
    make_scorer,
    root_mean_squared_error
)

def get_metrics(y_test: pd.Series, y_pred: np.ndarray) -> None:
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = root_mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    
    print(f"MAE {round(mae, 2)}")
    print(f"MSE {round(mse, 2)}")
    print(f"R2 {round(r2, 2)}")
    print(f"MAPE {round(mape, 2)}")
    print(f"RMSE: {round(rmse, 2)}")

def train_model(model: type[BaseEstimator], **kwargs: Any) -> None:
    # Instanciar el modelo
    model_instance = model(**kwargs)
    
    # Validación cruzada ANTES del entrenamiento final
    print("=== VALIDACIÓN CRUZADA ===")
    scorers = {
        'MAE': 'neg_mean_absolute_error',
        'RMSE': 'neg_root_mean_squared_error',
        'MSE': 'neg_mean_squared_error', 
        'R2': 'r2',
        'MAPE': make_scorer(mean_absolute_percentage_error, greater_is_better=False)
    }
    
    cv_results = cross_validate(
        model_instance, X_train, y_train,
        cv=5,
        scoring=scorers
    )
    
    # Mostrar métricas de CV
    for metric_name, scores in cv_results.items():
        if metric_name.startswith('test_'):
            clean_name = metric_name.replace('test_', '')
            if clean_name in ['MAE', 'MSE', 'MAPE']:
                cv_scores = -scores  # Convertir a positivo
            else:
                cv_scores = scores
            mean_score = np.mean(cv_scores)
            std_score = np.std(cv_scores)
            print(f"CV {clean_name}: {round(mean_score, 2)} (+/- {round(std_score * 2, 2)})")
    
    print("\n=== MÉTRICAS EN TEST SET ===")
    
    # Tu lógica original
    model_instance.fit(X_train, y_train)
    y_pred = model_instance.predict(X_test)
    
    return get_metrics(y_test, y_pred)
```
