```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv("/home/tomas/Escritorio/Proyectos/housing-prices/app/data/train.csv")
dataset
```

```python
dataset.set_index("Id", inplace=True)
dataset
```

```python
dataset.isna().sum()[dataset.isna().sum() > 0].sort_values(ascending=False)
```

```python
dataset.drop(columns=["PoolQC", "MiscFeature", "Alley", "Fence", "MasVnrType", "FireplaceQu", "LotFrontage"], inplace=True)
```

```python
dataset[dataset["GarageType"].isna()]
```

```python
dataset["GarageType"] = dataset["GarageType"].fillna("NoGarage")
```

```python
dataset["GarageYrBlt"] = dataset["GarageYrBlt"].fillna(dataset["YearBuilt"])
```

```python
dataset["GarageFinish"] = dataset["GarageFinish"].fillna(dataset["YearBuilt"])
```

```python
dataset["GarageQual"] = dataset["GarageQual"].fillna("NQ")
```

```python
dataset["GarageCond"] = dataset["GarageCond"].fillna("NQ")
```

```python
dataset.drop(columns=["MSZoning", "Street", "LandContour", "Utilities", "LandSlope", "Condition2", "RoofMatl", "BsmtCond", "BsmtFinType2", "Heating", "CentralAir", "Electrical", "LowQualFinSF", "Functional", "GarageFinish", "GarageQual", "GarageCond", "PavedDrive", "SaleType"], inplace=True)
```

```python
dataset["TotalBsmtSF"] = dataset["TotalBsmtSF"]-dataset["BsmtUnfSF"]
dataset
```

```python
dataset.drop(columns=["BsmtFinSF1", "BsmtFinSF2", "BsmtUnfSF"], inplace=True)
```

```python
dataset["2ndFlrSF"] = dataset["2ndFlrSF"].apply(lambda x:True if x > 0 else False)
dataset["2ndFlrSF"] = dataset["2ndFlrSF"].astype(int)
```

```python
dataset.drop(columns=["1stFlrSF"],inplace=True)
```

```python
dataset["TotalBath"] = dataset["FullBath"]+dataset["HalfBath"]
```

```python
dataset.drop(columns=["BsmtFullBath", "BsmtHalfBath", "FullBath", "HalfBath"],inplace=True)
```

```python
dataset["OutdoorArea"] = dataset["WoodDeckSF"] + dataset["OpenPorchSF"] + dataset["EnclosedPorch"] + dataset["3SsnPorch"] + dataset["ScreenPorch"]
```

```python
dataset.drop(columns=["WoodDeckSF", "OpenPorchSF", "EnclosedPorch", "3SsnPorch", "ScreenPorch"],inplace=True)
```

```python
dataset.drop(columns=["GarageType", "GarageYrBlt"],inplace=True)
```

```python
dataset["TotalArea"] = dataset["GrLivArea"]+dataset["TotalBsmtSF"]+dataset["GarageArea"]+dataset["OutdoorArea"]
```
